```python
# imports
import pickle
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('/content/drive/MyDrive/Datasets/risco_credito.pkl', 'rb') as f:
  X_risco, y_risco = pickle.load(f)
```

```python
X_risco
```

```python
y_risco
```

```python
clf = DecisionTreeClassifier(criterion='entropy').fit(X_risco, y_risco)
```

```python
print(f"Hitórico de crédito: {(clf.feature_importances_[0])*100}")
print(f"Dívida: {(clf.feature_importances_[1])*100}")
print(f"Garantia: {(clf.feature_importances_[2])*100}")
print(f"Renda: {(clf.feature_importances_[3])*100}")
```

```python
predict = ['história', 'dívida', 'garantias', 'renda']
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,10))
tree.plot_tree(clf, feature_names=predict, class_names=clf.classes_, filled=True);
```

```python
# visualizando previsões
prev = clf.predict([[0,0,1,2],[2,0,0,0]])
prev
```
