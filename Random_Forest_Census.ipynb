```python
# imports
import pickle as pkl
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
from sklearn import tree
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('/content/drive/MyDrive/Datasets/census.pkl', 'rb') as f:
  X_train, y_train, X_test, y_test = pkl.load(f)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
clf = RandomForestClassifier(n_estimators=99 ,criterion='entropy',
                             max_depth=8,random_state=42).fit(X_train, y_train)
prediction = clf.predict(X_test)
```

```python
accuracy_score(y_test, prediction)
```

```python
conf_matrix = ConfusionMatrix(clf).fit(X_train, y_train)
```

```python
conf_matrix.score(X_test, y_test)
```

```python
print(classification_report(y_test, prediction))
```
